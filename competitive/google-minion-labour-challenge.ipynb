```python
ls=[1, 2, 2, 3, 3, 3, 4, 5, 5]
n=1
if(len(ls)>100):
    raise Exception('Items exceeds the maximum allowed length of 100')
set2=countValue(ls,n) 
ls2=list(set2) 
print(list(filter(lambda x: x not in ls2, ls)))
```

```python
def countValue(ls,n):
    my_set=set()
    for i in ls:
        if ls.count(i) > n:
            my_set.add(i)
    return my_set 
```
